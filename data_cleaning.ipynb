```python
import pandas as pd
import json
```

```python
column_names = ["ID", "Category", "Attribute1", "Attribute2", "Flag", "Section", "Policy_JSON", "Date", "URL"]
data = pd.read_csv("./OPP-115/annotations/20_theatlantic.com.csv", header=None, names=column_names)

def parse_policy_entry(entry):
    try:
        entry_dict = json.loads(entry)
        
        other_type_text = entry_dict.get("Other Type", {}).get("selectedText", "")
        other_type_value = entry_dict.get("Other Type", {}).get("value", "Unspecified")
        
        collection_mode = entry_dict.get("Collection Mode", {}).get("value", "Unspecified")
        collection_text = entry_dict.get("Collection Mode", {}).get("selectedText", "")
        
        choice_scope = entry_dict.get("Choice Scope", {}).get("value", "Unspecified")
        choice_text = entry_dict.get("Choice Scope", {}).get("selectedText", "")
        
        action_type = entry_dict.get("Action First-Party", {}).get("value", "Unspecified")
        action_text = entry_dict.get("Action First-Party", {}).get("selectedText", "")
        
        personal_info_type = entry_dict.get("Personal Information Type", {}).get("value", "Unspecified")
        personal_info_text = entry_dict.get("Personal Information Type", {}).get("selectedText", "")
        
        identifiability = entry_dict.get("Identifiability", {}).get("value", "Unspecified")
        identifiability_text = entry_dict.get("Identifiability", {}).get("selectedText", "")
        
        does_or_not = entry_dict.get("Does/Does Not", {}).get("value", "Unspecified")
        does_or_not_text = entry_dict.get("Does/Does Not", {}).get("selectedText", "")
        
        user_type = entry_dict.get("User Type", {}).get("value", "Unspecified")
        purpose = entry_dict.get("Purpose", {}).get("value", "Unspecified")

        return (
            other_type_text, other_type_value, collection_mode, collection_text, 
            choice_scope, choice_text, action_type, action_text, 
            personal_info_type, personal_info_text, identifiability, 
            identifiability_text, does_or_not, does_or_not_text, user_type, purpose
        )
    except json.JSONDecodeError:
        return ("Unspecified",) * 16
data[
    [
        'other_type_text', 'other_type_value', 'collection_mode', 'collection_text', 
        'choice_scope', 'choice_text', 'action_type', 'action_text', 
        'personal_info_type', 'personal_info_text', 'identifiability', 
        'identifiability_text', 'does_or_not', 'does_or_not_text', 'user_type', 'purpose'
    ]
] = data['Policy_JSON'].apply(parse_policy_entry).apply(pd.Series)
print(data.head(10))
```
